```python
def plot(df):
    from bokeh.io import output_file, show
    from bokeh.models import ColumnDataSource
    from bokeh.palettes import Spectral6
    from bokeh.plotting import figure,output_notebook
    from bokeh.transform import factor_cmap

    output_notebook()

    Attributes = df.columns.values
    counts = df.iloc[0].values

    source = ColumnDataSource(data=dict(Attributes=Attributes, counts=counts))

    p = figure(x_range=Attributes, plot_height=650, toolbar_location=None, title="Attribute Counts")
    p.vbar(x='Attributes', top='counts', width=0.9, source=source, legend_field="Attributes",
           line_color='white', fill_color=factor_cmap('Attributes', palette=Spectral6, factors=Attributes))

    p.xgrid.grid_line_color = None
    p.y_range.start = 0
    p.y_range.end = 500
    p.legend.orientation = "horizontal"
    p.legend.location = "top_center"

    show(p)
    
```

```python
covid19_dict = {
    'Country' : ['USA','China','Italy','Spain'],
    'Cases' : [141559,97689,81439,80110],
    'Recovered': ['4,435','13,030','75,448','14,709'],
}
```

```python
data_covid = pd.DataFrame(data = covid19_dict)
```

```python
data_covid
```

```python
req = data_covid[['Country','Cases']]
```

```python
req.rename(columns={'Cases':'value','Country':'country'})
```

```python
def plot(req):
    from math import pi
    from bokeh.io import output_file, show
    from bokeh.palettes import Category20c
    from bokeh.plotting import figure, output_notebook
    from bokeh.transform import cumsum
    
    output_notebook()
    x = req
    print(x)


    data = req.rename(columns={'Cases':'value','Country':'country'})
    data['angle'] = data['value']/data['value'].sum() * 2*pi
    data['color'] = Category20c[len(x)]

    p = figure(plot_height=350, title="Pie Chart", toolbar_location=None,
               tools="hover", tooltips="@country: @value", x_range=(-0.5, 1.0))

    p.wedge(x=0, y=1, radius=0.4,
            start_angle=cumsum('angle', include_zero=True), end_angle=cumsum('angle'),
            line_color="white", fill_color='color', legend_field='country', source=data)

    p.axis.axis_label=None
    p.axis.visible=False
    p.grid.grid_line_color = None

    show(p)
    
```

```python
plot(req)
```
